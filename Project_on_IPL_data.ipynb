```python
bat_first = ipl[ipl['win_by_runs']!=0]
print(bat_first[['win_by_runs']].count())
```

```python
bat_first = ipl[ipl['win_by_runs']!=0]
print(bat_first[['winner','win_by_runs']].to_string())
```

```python
plt.figure(figsize=(5,5 ))
plt.hist(bat_first['win_by_runs'],color="purple")
plt.title("Distribution of Runs")
plt.xlabel("Runs")
plt.show()
```

```python
bat_first['winner'].value_counts()
```

```python
plt.figure(figsize=(6,6))
plt.bar(list(bat_first['winner'].value_counts()[:3].keys()),
        list(bat_first['winner'].value_counts()[:3]),
             color=["blue","yellow","red"])
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(bat_first['winner'].value_counts()),
         labels=list(bat_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
